# Dados Outubro 2003

```python
import numpy as np
import glob
import pylab
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
print(glob.glob('vss200310[0-30]*dmin.min')) #listagem de arquivos
```

* **Empilhando os valores listados por minuto para todos os dias do mÃªs**

```python
filenames = glob.glob('vss200310*dmin.min')    


for f in filenames:                    
    
    data = np.loadtxt(fname=f, skiprows=26, usecols= (3,4,5,6))
    
    if f == filenames[0]:
        h_out = data[:,0]
        f_out = data[:,3]
        
    else:
        h = data[:,0]
        h_out = np.hstack((h_out,h))
        f = data[:,3]
        f_out = np.hstack((f_out,f))
        
print(h_out)
print(f_out)

h_out[h_out > 60000] = np.nan 
print(h_out)

f_out[f_out > 60000] = np.nan 
print(f_out)
```

* **Plot para a componente H(nT) em Outubro de 2003**

```python
d = h_out           
N = h_out.shape[0]
time = range(0,N)

fig = plt.figure(figsize=(15.0, 5.0))

H = plt.plot(time, d)

plt.xticks([d*1440 for d in range(31)], [d+1 for d in range(31)])

plt.ylabel('H(nT)')
plt.xlabel('Day')
plt.show()
```

* **Plot para a componente F(nT) em Outubro de 2003**

```python
d = f_out           
N = f_out.shape[0]
time = range(0,N)

fig = plt.figure(figsize=(15.0, 5.0))

F = plt.plot(time, d)

plt.xticks([d*1440 for d in range(31)], [d+1 for d in range(31)])

plt.axhline(y= 23100, linewidth=1, color='grey')
plt.axhline(y= 23200, linewidth=1, color='grey')
plt.axhline(y= 23300, linewidth=1, color='grey')
plt.axhline(y= 23400, linewidth=1, color='grey')
plt.axhline(y= 23500, linewidth=1, color='grey')

plt.ylabel('F(nT)')
plt.xlabel('Day')
plt.show()
```

* **Selecionado o range entre os dias 27 e 31 de outubro**
>Para a componente H

```python

d = h_out[37440:40650]

e = h_out[40650:44640]

N = h_out.shape[0]
time1 = range(37440,40650)

time2= range(40650,44640)

fig = plt.figure(figsize=(15.0, 10.0))

Sq = plt.plot(time1, d, label="Sq")

Storm = plt.plot(time2, e, label="Storm", color='red')
plt.xticks([dia*1440 for dia in range(26,32)], ['%i/10/2003'%dia for dia in range(27,32)], fontsize=17)
plt.yticks(np.arange(18700, 19500, 100), fontsize=18)
pylab.ylim([18670,19400])

plt.ylabel('H (nT)', fontsize=26)
plt.xlabel('Time [days]',fontsize=26)
plt.grid(True)
plt.legend(fontsize=18)
plt.savefig('H27_31_october(2003).png')

plt.show()
```
